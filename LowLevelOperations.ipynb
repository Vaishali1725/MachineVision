```python
import cv2
```

Reading & Displaying Image

```python
image = cv2.imread("/Users/vaishaliyadav/Desktop/lenna.png") # Read The Image
# cv2.imshow("Picture", image) # Frame Title with the image to be displayed 
# cv2.waitKey(5)
```

```python
import matplotlib.pyplot as plt 
plt.imshow(image);
```

Writing Image

```python
cv2.imwrite("lena1.png", image)
```

Resizing Image

```python
cv2.imshow("Picture", image)
resize = cv2.resize(image, (256,256))
# cv2.imshow("Picture", image)
# cv2.imshow("Resized Picture", resize) # Resized Image
# cv2.waitKey(0)
```

```python
plt.imshow(resize);
```

Converting Into Gray Scale Image

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Convert the color image to grayscale image
# cv2.imshow("Gray Picutre", gray) # Display Grayscale image
# cv2.waitKey(0)
```

```python
plt.imshow(gray);
```

Blurring the Image

```python
blur = cv2.GaussianBlur(gray, (19,19), 0) 
# cv2.imshow("Blur Picutre", blur) # Display Blur image
# cv2.waitKey(0)
```

```python
plt.imshow(blur);
```

Crop an Image

```python
face = image[85:250, 85:220]
# cv2.imshow("Face", face)
# cv2.waitKey(0)
```

```python
plt.imshow(face);
```

Drawing Lines

```python
import numpy as np
```

```python
img = np.zeros((512,512,3))
img = cv2.line(img,(0,0),(511,511),(255,0,0),10)

# cv2.imshow("Line", img) # Display the line cutting through the image
# cv2.waitKey(0)
```

```python
plt.imshow(img)
```

Drawing Rectangle

```python
img = np.zeros((512,512,3))
img = cv2.rectangle(img,(200,200),(300,300),(0,0,255),10)

# cv2.imshow("Rectangle", img) # Display the rectangle
# cv2.waitKey(0)
```

```python
plt.imshow(img)
```

Drawing a Circle

```python
img = np.zeros((512,512,3))
img = cv2.circle(img,(255,255),100,(0,255,255),10)

# cv2.imshow("Circle", img) # Display the circle
# cv2.waitKey(0)
```

```python
plt.imshow(img)
```

Inserting Text

```python
img = np.zeros((1024,1024,3))
font = cv2.FONT_HERSHEY_SIMPLEX 
mat = cv2.putText(img, 'Computer Vision Class 1', (125,250), 
            font, 1, (0, 0, 255), 3, cv2.LINE_AA) 

# cv2.imshow("TEXT", img) # Display the TEXT
# cv2.waitKey(0)
```

```python
plt.imshow(img)
```

Drawing Polygons

```python
img = np.zeros((512,512,3))

pts = np.array([[25, 70], [25, 160],  
                [110, 200], [200, 160],  
                [200, 70], [110, 20]], 
               np.int32) 

pts=pts.reshape((-1,1,2))
mat = cv2.polylines(img,[pts],True,(0,255,255))

# cv2.imshow("Polygon", img) # Display the Hexagon
# cv2.waitKey(0)
```

```python
plt.imshow(img)
```
